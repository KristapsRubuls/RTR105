```python
from numpy import *
from matplotlib import pyplot as plt
```

```python
x = linspace(0, 11, 80)
```

```python
y = sinh(sqrt(x))
y1 = sqrt(x)*(pow(x, 0))/(2*0+1)
y2 = sqrt(x)*(pow(x, 0))/(2*0+1) + x*sqrt(x)/6
y3 = sqrt(x)*(pow(x, 0))/(2*0+1) + x*sqrt(x)/6 + pow(x, 2)*sqrt(x)/(5*4*3*2)
y4 = sqrt(x)*(pow(x, 0))/(2*0+1) + x*sqrt(x)/6 + pow(x, 2)*sqrt(x)/(5*4*3*2) + pow(x, 3)*sqrt(x)/(7*6*5*4*3*2)
```

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Funkcija sinh(sqrt(x)) un tas izvirzijumi")
plt.plot(x, y)
plt.plot(x, y, color = "#00FF00")
plt.plot(x, y1)
plt.plot(x, y1, color = "#0FFFF0")
plt.plot(x, y2)
plt.plot(x, y2, color = "#0000FF")
plt.plot(x, y3)
plt.plot(x, y3, color = "#FF0000")
plt.plot(x, y4)
plt.plot(x, y4, color = "#FFFF00")
plt.show()
```
